```python
import random


def genera_quadrato_latino(n):
    quadrato = [[0] * n for _ in range(n)]

    # Inizializza la prima riga con numeri casuali
    riga_iniziale = list(range(1, n + 1))
    random.shuffle(riga_iniziale)
    quadrato[0] = riga_iniziale

    # Costruisce il resto del quadrato
    for riga in range(1, n):
        for colonna in range(n):
            numero = quadrato[0][(colonna + riga) % n]
            quadrato[riga][colonna] = numero

    return quadrato
```

```python
dimensione = int(input("Inserisci le dimensioni del quadrato latino: "))
quadrato_latino = genera_quadrato_latino(dimensione)

# Stampa il quadrato latino
for riga in quadrato_latino:
    print(riga)
```
